```python
import pandas as pd
import glob
import os
import re

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# First DataFrame
df1 = pd.read_csv('Bleifrei_TS_raw_excel.csv',
                 sep=';',
                 encoding='latin-1')
  
# Second DataFrame
df2 = pd.read_csv('Diesel_TS_raw_excel.csv',
                 sep=';',
                 encoding='latin-1')
  
# append method
merged = df1.append(df2)
display(merged)
```

```python
#delet usless data

merged.drop(columns=['web-scraper-order', 'web-scraper-start-url'], inplace=True, axis=1)

merged
```

```python
# Extract values from 'price_raw' strings
price = []
for i in merged['Preis']:
    d1 = re.findall('[0-9]+', i)
    try:
        d2 = d1[0].strip()
    except:
        d2 = None
    price.append(d2)

# Save as new variable in the pandas data frame
merged['price'] = pd.Series(price)

# Print first 5 values
print(merged['Preis'].head(5), '\n')
print(merged['price'].head(5))
```

```python
# Extract values from 'price_raw' strings
price = []
for i in merged['Preis2']:
    d1 = re.findall('[0-9]+', i)
    try:
        d2 = d1[0].strip()
    except:
        d2 = None
    price.append(d2)

# Save as new variable in the pandas data frame
merged['price2'] = pd.Series(price)

# Print first 5 values
print(merged['Preis2'].head(5), '\n')
print(merged['price2'].head(5))
```

```python
#export Data

merged.to_csv('combined.csv',
                 sep=';',
                 encoding='latin-1',
                 index=False)
```
